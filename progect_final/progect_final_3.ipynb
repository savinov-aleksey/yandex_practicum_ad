### Для каждой книги посчитайте количество обзоров и среднюю оценку

```python
query = '''

SELECT books.book_id, 
       books.title, 
       COUNT(DISTINCT reviews.review_id) AS count_reviews, 
       ROUND(AVG(ratings.rating),1) AS avg_rating
FROM books

LEFT JOIN reviews
ON books.book_id = reviews.book_id

LEFT JOIN  ratings
ON books.book_id = ratings.book_id

GROUP BY books.book_id, books.title
ORDER BY count_reviews DESC

'''

pd.io.sql.read_sql(query, con = engine)
```

Получили данные на 1000 книг с количеством обзоров от 0 до 7 и средним рейтингом. Количество обзоров не влияет на рейтин книг, при 0 обзоров рейтинг книги может ровняться 5.

### Определите издательство, которое выпустило наибольшее число книг толще 50 страниц — так вы исключите из анализа брошюры

```python
query = '''

WITH filter AS (SELECT publisher_id, book_id 
                FROM books 
                WHERE num_pages > 50)

SELECT filter.publisher_id, publishers.publisher
FROM filter 

INNER JOIN publishers
ON filter.publisher_id = publishers.publisher_id

GROUP BY filter.publisher_id, publishers.publisher
ORDER BY COUNT(filter.book_id) DESC

LIMIT 1

'''

pd.io.sql.read_sql(query, con = engine)
```

Сгруппированные и отсортированные по убыванию данные показывают издательство выпустившее наибольшее количество книг **Penguin Books**.

### Определите автора с самой высокой средней оценкой книг — учитывайте только книги с 50 и более оценками

```python
query = '''

WITH 

filter AS (SELECT book_id
            FROM ratings
            GROUP BY book_id
            HAVING COUNT(rating_id) > 50)


SELECT authors.author_id, 
       authors.author, 
       AVG(ratings.rating) AS avg_rating
FROM books

INNER JOIN filter
ON books.book_id = filter.book_id

INNER JOIN authors
ON books.author_id = authors.author_id

INNER JOIN ratings
ON filter.book_id = ratings.book_id

GROUP BY authors.author_id, authors.author
ORDER BY AVG(ratings.rating) DESC

LIMIT 1

'''

pd.io.sql.read_sql(query, con = engine)
```

### Посчитайте среднее количество обзоров от пользователей, которые поставили больше 48 оценок

```python
query = '''

SELECT COUNT(DISTINCT reviews.text)/ COUNT(DISTINCT reviews.username) AS avg_reviews_count
FROM ratings

INNER JOIN reviews
ON ratings.username = reviews.username

WHERE ratings.username in ( SELECT username
                            FROM ratings
                            GROUP BY username
                            HAVING COUNT(rating_id) > 48)

'''

pd.io.sql.read_sql(query, con = engine)
```

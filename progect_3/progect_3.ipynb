## Шаг 2. Предобработка данных

Привожу название колонок к единообразию.

```python
df = df.rename(columns = {'cityCenters_nearest': 'city_centers_nearest',
                          'parks_around3000' : 'parks_around_3000',
                         'ponds_around3000' : 'ponds_around_3000'})
```

### Найдите и изучите пропущенные значения в столбцах:

- Определите, в каких столбцах есть пропуски.

Процент пропущеных значений в колонках.

```python
round(df.isna().sum()/df.shape[0]*100, 1)
```

Видно что в датасете много колонок с пропусками. Исходя из названий колонок предположу,что не все пропуски нужно будет заполнить. В колонках `ponds_nearest, parks_nearest` оставлю пустые значения, вероятно такие квартиры не расположены в близком расстоянии от парков и прудов.

- Заполните пропущенные значения там, где это возможно. Например, если продавец не указал число балконов, то, скорее всего, в его квартире их нет. Такие пропуски правильно заменить на 0. Если логичную замену предложить невозможно, то оставьте эти значения пустыми. Пропуски — тоже важный сигнал, который нужно учитывать.

```python
(df[['balcony',
    'parks_around_3000',
    'ponds_around_3000']]) = (df[['balcony',
                              'parks_around_3000', 
                              'ponds_around_3000']].fillna(0))
```

Пердполагаю что не заполнение данных в столбцах `balcony`,`parks_around3000`, `ponds_around3000` предполагает 0. Столбец is_apartment имеет булевые значения, не заполненные значения меняю на `False`. Предполагаю, что причиной не заполнения данных столбцов был человеческий фактор, отказ или пропуск в заполнении.

```python
df['is_apartment'] = df['is_apartment'].fillna(False)
```

В столбце `locality_name` удаляю не заполненные строки ввиду малого их количества они не окажут значимого влияния на иследование.

```python
print(f"всего значений {df['locality_name'].count()},пустых строк {df['locality_name'].isna().sum()}")
df.dropna(subset=['locality_name'], inplace=True)
df = df.reset_index(drop=True)
```

Пустые значения в колонке `ceiling_height` заполняю медианным значением. Явно ошибочные значения меняю на логично подходящие, причины ошибок считаю человеческий фактор, опечатки. Значения менее 2 метров и более 4 метров заменяю на медианное и среднее значение.


```python
list(df['ceiling_height'].sort_values(ascending=True).unique())
```

```python
ceiling_height_median = df['ceiling_height'].median()
df['ceiling_height'].fillna(ceiling_height_median, inplace=True)
```

Пропуски в колонках `kitchen_area` и `living_area` заполняю медианным значением из сгрупированных данных по количеству комнат в квартире.

```python
df['kitchen_area'] = df['kitchen_area'].fillna(df.groupby('rooms')['kitchen_area'].transform('median'))
```

```python
df['living_area'] = df['living_area'].fillna(df.groupby('rooms')['living_area'].transform('median'))
```

Оставшиеся пустые значения заполняю медианным значением от всего столбца данных.

Заполняю пропуски в стобцах `airports_nearest` и `cityCenters_nearest` средним значением расстояния сгрупированным по месторасположению квартиры.

```python
groupby_airpots_nearest = df.groupby('locality_name')['airports_nearest'].transform('mean')
df['airports_nearest'] = df['airports_nearest'].fillna(groupby_airpots_nearest)
```

Оставшиеся пустые значения заполняю средним значением от всего столбца данных.

```python
df['airports_nearest'] = df['airports_nearest'].fillna(df['airports_nearest'].mean())
```

```python
groupby_city_centers_nearest = df.groupby('locality_name')['city_centers_nearest'].transform('mean')
df['city_centers_nearest'] = df['city_centers_nearest'].fillna(groupby_city_centers_nearest)
```

Оставшиеся пустые значения заполняю средним значением от всего столбца данных.

```python
df.isna().sum()
```

### Изучите уникальные значения в столбце с названиями и устраните неявные дубликаты. Например, «поселок Рябово» и «поселок городского типа Рябово», «поселок Тельмана» и «посёлок Тельмана» — это обозначения одних и тех же населённых пунктов. Вы можете заменить названия в существующем столбце или создать новый с названиями без дубликатов.

Определяю уникальные значения в столбце `locality_name`.

```python
list(df['locality_name'].unique())
```

В столбце присутствует множество неявных дубликатов. Для придания единообразия оставим только то, что начинается с большой буквы во всем столбце.

```python
def up(string):
    for i in range(len(string)):
        if string[i].isupper():
            return string[i: len(string)]
```

```python
df['locality_name'] = df['locality_name'].apply(up)
```

```python
df['ceiling_height'] = df['ceiling_height'].replace({26 : 2.6,
                                                    27 : 2.7,
                                                    32 : 3.2,
                                                    25 : 2.5,
                                                    24 : 2.4,
                                                    27.5 : 2.75})
```
